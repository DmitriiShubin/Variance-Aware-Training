```python
patients = [i for i in os.listdir("../data/HepaticVessel/Task08_HepaticVessel/imagesTr/") if i.find('._h')==-1]
patients.sort()
```

```python
names = []
for patient in patients:
    img = nib.load("../data/HepaticVessel/Task08_HepaticVessel/imagesTr/"+patient)
    img = np.array(img.dataobj)
    seg = nib.load("../data/HepaticVessel/Task08_HepaticVessel/labelsTr/"+patient)
    seg = np.array(seg.dataobj)
    
    os.makedirs("../data/HepaticVessel/processed_data/"+patient[:-7], exist_ok=True)
    
    for i in range(img.shape[2]):
        temp = img[:,:,i]
        temp = temp.reshape(temp.shape[0],temp.shape[1],1)
        
        temp_seg = seg[:,:,i]
        temp_seg = temp_seg.reshape(temp_seg.shape[0],temp_seg.shape[1],1)
        
        name = "../data/HepaticVessel/processed_data/"+patient[:-7]+'/'
        name_seg = '_'.join(name.split('_')[:-1])+'_'+name.split('_')[-1]+f'{i}_seg.npy'
        name = '_'.join(name.split('_')[:-1])+'_'+name.split('_')[-1]+f'{i}.npy'
        
        
        np.save(name,temp)
        np.save(name_seg,temp_seg)
    
```
