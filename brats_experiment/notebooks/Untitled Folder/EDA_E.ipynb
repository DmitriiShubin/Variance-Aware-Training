```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
import os
import gc
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def find_disease(labels,disease_id,disease_array):
    
    for index,i in enumerate(labels):
        if disease_id in i:
            disease_array[index] = 1
            
    return disease_array
```
